## 1-4. 데이터 핸들링 - 판다스

판다스: 데이터 처리를 위한 라이브러리. 행과 열로 이뤄진 2차원 데이터를 효율적으로 가공 처리할 수 있는 기능 제공. 파이썬의 리스트, 컬렉션, 넘파이 등의 내부 데이터 뿐만 아니라 CSV 등의 파일을 쉽게 DataFrame 으로 변경해 데이터 가공 분석을 편리하게 만듦.
  - 핵심 객체: DataFrame(칼럼이 여러 개인 데이터 구조체)


### 판다스 시작 - 파일을 DataFrame으로 로딩, 기본 API
``` import pandas as pd ```
판다스 모듈 임포트

데이터 파일을 판다스의 DataFrame으로 로딩. read_csv(,), read_table(tab), read_fwf(Fixed width). Filepath에는 로드하려는 데이터 파일 경로를 포함한 파일명 입력.
- pd.read_csv(): 호출 시 파일명 인자로 들어온 파일을 로딩해 DataFrame 객체로 반환. 데이터 파일의 첫 번째 줄에 있던 칼럼 문자열이 DataFrame의 칼럼으로 할당. 별다른 파라미터 지정이 없으면 파일의 맨 처음 로우를 칼럼명으로 인지하고 칼럼으로 변환. 인덱스가 고유의 레코드를 식별하는 역할을 함.
- DataFrame.head(): DataFrame의 맨 앞에 있는 N개의 로우 반환
- DataFrame.shape(): DataFrame의 행과 열 크기 튜플 형태로 반환
- DataFrame.info(): 총 데이터 건수, 데이터 타입, Null 건수
- DataFrame.describe(): 칼럼별 숫자형 데이터값의 n-percentile 분포도, 평균값, 최댓값, 최솟값 나타냄. 오직 숫자형 칼럼의 분포도만 조사하며 자동으로 object 타입의 칼럼은 출력에서 제외. 개략적 수준의 분포도 확인 가능.
- count(): Not Null인 데이터 건수
- mean(): 전체 데이터의 평균값
- std(): 표준편차, min(): 최솟값, max(): 최댓값
- 25%: 25 percentile 값, 50%: 50 percentile 값, 75%: 75 percentile 값
- value_counts(): 해당 칼럼값의 유형과 건수 확인. 데이터 분포도 확인에 유용한 함수
- DataFrame[칼럼]: 해당 칼럼에 해당하는 Series 객체 반환
- Series: Index와 단 하나의 칼럼으로 구성된 데이터 세트

### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환
DataFrame은 파이썬의 리스트, 딕셔너리, 넘파이 ndarray 등 다양한 데이터로부터 생성되고 상호 변환 가능

#### 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기
DataFrame은 칼럼명으로 인하여 리스트와 넘파이 ndarray 보다 상대적으로 편하게 데이터 핸들링 가능. DataFrame으로 변환 시 칼럼명 지정. DataFrame 객체의 생성인자 data는 리스트나 딕셔너리 또는 넘파이 ndarray를 입력받고, 생성인자 columns는 칼럼명 리스트를 입력 받아 쉽게 DataFrame 생성. 2차원 이하의 데이터들만 DataFrame으로 변환 간으.
- 1차원 데이터 DataFrame 생성: 칼럼명이 1개만 필요
- 2차원 데이터 DataFrame 생성: 리스트와 ndarray를 기반으로 칼럼명 여러 개 필요
- 딕셔너리를 DataFrame으로 변환: 딕셔너리의 키는 칼럼명, 딕셔너리의 값은 키에 해당하는 칼럼 데이터로 변환. 키의 경우는 문자열, 값의 경우 리스트 형태로 딕셔너리 구성

#### DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기
- tolist():  DataFrame을 리스트로 변환
- to_dict(): DataFrame을 딕셔너리로 변환

### DataFrame의 칼럼 데이터 세트 생성과 수정
- 생성: DataFrame[] 내에 새로운 칼럼명 입력 값 할당. 칼럼 Series에 값 할당하고 DataFrame에 추가. Series에 상숫값 할당하면 Series의 모든 데이터 세트에 일괄 적용.
- 수정: 업데이트를 원하는 칼럼 Series를 DataFrame[] 내에 칼럼명으로 입력 후 값 할당.

### DataFrame 데이터 삭제
- drop(): 데이터 삭제

``` DataFrame.drop(labels=None, axis=0, index=None, columns=None, inplace=False, errors='raise')```
  - axis: axis 0: 로우 방향 축, axis 1: 칼럼 방향 축. drop(axis=1) 칼럼 드롭. drop(axis=0) 로우 드롭
  - inplace: inplace=False 자기 자신의 DataFrame의 데이터는 삭제하지 않으며 삭제된 결과 DataFrame을 반환. inplace=True 자신의 DataFrame의 데이터를 삭제.
  - 여러 개의 칼럼 삭제하고 싶으면 리스트 형태로 삭제하고자 하는 칼럼 명을 입력해 labels 파라미터로 입력
  - inplace=True 하면 반환 값이 None이 됨.

### Index 객체
- Index: DataFrame, Series의 레코드를 고유하게 식별하는 객체. DataFrame.index, Series.index
- 식별성 데이터를 1차원 array로 가지고 있음. ndarray와 유사하게 단일 값 반환 및 슬라이싱
- 한 번 만들어진 DataFrame 및 Series의 Index 객체는 함부로 변경 불가. Index는 오직 식별용으로 사용
- reset_index(): 새롭게 인덱스를 연속 숫자 형으로 할당하며 기존 인덱스는 index라는 새로운 칼럼명으로 추가 Series에 reset_index() 적용하면 DataFrame이 반환. drop=True로 설정하면 기존 인덱스는 새로운 칼럼으로 추가되지 않고 삭제됨.

### 데이터 셀렉션 및 필터링
- 넘파이: [] 연산자 내 단일 값 추출, 슬라이싱, 팬시 인덱싱, 불린 인덱싱을 통해 데이터 추출
- 판다스: ix[], iloc[], loc[] 연산자 통해 동일 작업 수행

#### DataFrame의 [] 연산자
- DataFrame[]: 행의 위치, 열의 위치, 슬라이싱 범위 등을 지정해 데이터 가져옴. 칼럼 지정 연산자
- 판다스의 인덱스 형태로 변환 가능한 표현식 입력 가능. 슬라이싱 이용
- 불린 인덱싱 표현 가능. 원하는 데이터를 편리하게 추출

#### DataFrame ix[]연산자
- 행 위치 지정으로 인덱스값 0, 열 위치 지정으로 원하는 위치의 데이터 추출 가능. 칼럼 위치 기반 인덱싱, 칼럼 명칭 기반 인덱싱
- 단일 지정, 슬라이싱, 불린 인덱싱, 팬시 인덱싱 가능

#### 명칭 기반 인덱싱과 위치 기반 인덱싱의 구분
- 명칭 기반 인덱싱: 칼럼의 명칭을 기반으로 위치를 지정하는 방식.
- 위치 기반 인덱싱: 0을 출발점으로 하는 가로축, 세로축 좌표 기반의 행과 열 위치를 기반으로 데이터 지정. 행, 열 값으로 정수 입력
- DataFrame의 인덱스값은 명칭 기반 인덱싱으로 간주

#### DataFrame iloc[] 연산자
- iloc[]: 위치 기반 인덱싱만 허용. 행 열 값으로 정수 또는 정수 형의 슬라이싱, 팬시 리스트 값을 입력해야 함. 명칭을 입력하면 오류 발생. 문자열 인덱스를 행 위치에 입력해도 오류 발생. 불린 인덱싱은 제공하지 않음

#### DataFrame loc[] 연산자
- loc[]: 명칭 기반 데이터 추출. 행 위치에는 DataFrame index 값, 열 위치에는 칼럼명 입력. index가 숫자 형일 수 있음.슬라이싱 기호를 적용하면 종료 값까지 포함.

#### 불린 인덱싱
- 자동으로 원하는 값 필터링.
- and &, or |, not ~

### 정렬, Aggregation 함수, GroupBy 적용

#### DataFrame, Series의 정렬 - sort_values()
- sort_values(): 주요 입력 파라미터는 by, ascending, inplace
  - by: 특정 칼럼을 입력하면 해당 칼럼으로 정렬 수행
  - ascending=True: 오름차순 정렬
  - ascending=False: 내림차순 정렬
  - inplace=False: 호출한 DataFrame 그대로 유지하며 정렬된 DataFrame 결과로 반환
  - inplace=True: 호출한 DataFrame 정렬 결과 그대로 적용.

#### Aggregation 함수 적용
- DataFrame에서 바로 aggregation 호출할 경우 모든 칼럼에 해당 aggregation 적용.
- 특정 칼럼에 aggregation 함수를 적용하기 위해서는 DataFrame에 대상 칼럼들만 추출해 aggregation 적용.

#### groupby() 적용
- 입력 파라미터 by 에 칼럼을 입력하면 대상 칼럼으로 groupby 됨.
- DataFrame에 groupby()를 호출하면 DataFrameGroupBy라는 또 다른 형태의 DataFrame 반환.
- groupby()를 호출해 반환된 결과에 aggregation 함수를 호출하면 groupby() 대상 칼럼을 제외한 모든 칼럼에 해당 aggregation 함수 적용
- 여러 개의 aggregation 함수명을 DataFrameGroupBy 객체의 agg() 내에 인자로 입력해서 사용

### 결손 데이터 처리하기
- 결손 데이터: 칼럼에 값이 없는 NULL인 경우. NaN으로 표시
- NaN 값은 평균 총합 등의 함수 연산 시 제외 됨.

#### isna()로 결손 데이터 여부 확인
- isna(): 데이터가 NaN인지 아닌지 알려줌. 모든 칼럼의 값이 NaN인지 아닌지 True False로 알려줌
- 결손 데이터 개수: isna() 결과에 sum() 함수를 추가해 구함.

#### fillna()로 결손 데이터 대체하기
- fillna(): 결손 데치터를 다른 값으로 대체. inplace=True 파라미터를 추가해야 실제 데이터 세트 값이 변경됨.

### apply lambda 식으로 데이터 가공
- apply 함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하는 기능 제공.
- lambda: 함수의 선언과 함수 내의 처리를 한 줄의 식으로 쉽게 변환하는 식, 여러 개의 값을 입력 인자로 사용해야 할 경우 map() 함수 결합 사용
- if , else 만 지원함.

````python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x:'Child' if x <= 15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
````

````python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x:'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
````

````python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat =''
  if age <= 5: cat='Baby'
  elif age<=12: cat = 'Child'
  elif age<=18: cat='Teenager'
  elif age<=25: cat='Student'
  elif age<=35: cat='Young Adult'
  elif age<=60: cat='Adult'
  else: cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_category(x)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age','Age_cat']].head()
````
