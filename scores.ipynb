````python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
df = pd.DataFrame(data=raw_data)
df
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = df['preTestScore'],
    y = df['postTestScore'],
    s = df['age'],# 점 size
    alpha = 0.5 # 점의 투명도
)
axes1.set_title('preTestScore vs postTestScore : Sized by Age')
axes1.set_xlabel('preTestScore')
axes1.set_ylabel('postTestScore')
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = df['preTestScore'],
    y = df['postTestScore'],
    s = df['postTestScore']*4.5,# 점 size
    c = df['female'],
    alpha = 0.5 # 점의 투명도
)
#axes1.legend(fontsize=12, loc='upper left')
axes1.set_title('preTestScore vs postTestScore : Sized by postTestScore, Colored by Sex')
axes1.set_xlabel('preTestScore')
axes1.set_ylabel('postTestScore')
````

### BONUS: Create your own question and answer it.


#### 1. Create a scatterplot of (postTestScore - preTestScore) and age and have the colors determined by the sexes and add a legend

````python
df['post-pre'] = df['postTestScore'] - df['preTestScore']
df
````

````python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = df['age'],
    y = df['post-pre'],
    c = df['female'],
)
axes1.set_title('(postTestScore - preTestScore) vs Age : Colored by Sex')
axes1.set_xlabel('Age')
axes1.set_ylabel('postTestScore - preTestScore')
axes1.legend(df['female'], loc='center right')
````

#### 2. Create a scatterplot of postTestScore against preTestScore and have the markers determined by the sexes.

````python
scatter = sns.lmplot(x = 'preTestScore', y = 'postTestScore',
                     data = df,
                     fit_reg = False,
                     hue = 'female',
                     markers = ['o', 'x']
                     )
scatter.set_axis_labels(xlabel='preTestScore', ylabel='postTestScore')
scatter.fig.suptitle('Scatterplot of preTestScore and postTestScore', fontsize=10, y=1.03)
````
