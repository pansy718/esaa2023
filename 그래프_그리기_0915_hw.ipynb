## 4-3. seaborn 라이브러리 자유자재로 사용하기

#### 실습 4. 단변량 그래프 그리기 - 히스토그램
1. seaborn 라이브러리 sns 로 줄여 사용
2. subplots, distplot 메서드 사용
3. distplot: 히스토그램과 밀접도 그래프 같이 그림. 밀집도 kde
4. 밀집도 그래프만 나타내려면 hist=False
5. rug: 양탄자 그래프. 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도 표현한 그래프
6. count 그래프: 이산값을 나타낸 그래프

#### 실습 5. 다양한 종류의 이변량 그래프 그리기
1. seaborn 라이브러리에서 산점도 그래프: regplot 메서드 사용. 산점도 그래프와 회귀선 함께 그림. 회귀선 제거: fit_reg=False
2. jointplot: 산점도 그래프와 히스토그램 한 번에 그려줌.
3. 육각 그래프: 2차원 표면에 육각형으로 데이터 쌓아 표현하는 그래프. jointplt 메서드 그대로 사용 kind='hex'
4. 이차원 밀집도 그리기: kdeplot 메소드 사용. shade=True: 그래프에 음영 효과
5. 바 그래프 그리기: 지정한 변수의 평균을 계산해 그림
6. 박스 그래프 그리기. 최솟값, 1분위수, 중간값, 3분위수, 최댓값, 이상치 등 다양한 통계량 한 번에 표현하는 그래프
7. 바이올린 그래프: 박스 그래프에 커널 밀도를 추정한 그래프
8. pairplot: 관계 그래프. 중복된 정보가 표현된다는 단점. 중복 그래프가 그려지는 위치를 직접지정해 원하는 그래프로 교체.
9. map_upper: 대각선 기준 위쪽에 그릴 그래프 지정 map_lower: 대각선 기준 아래에 그릴 그래프 지정. map_diag: 대각선 중심으로 그래프 그림

#### 실습 6. 다변량 그래프 그리기
1. seaborn 라이브러리로 바이올린 그래프 그리기 색상추가: hue
2. 산점도, 관계 그래프 그리기 색상 추가: hue
3. 산점도 그래프의 크기와 모양 조절하기 scatter_kws에 딕셔너리 형태로 인잣값 전달
4. 산점도 그래프의 점을 다른 기호로 표현하고 싶다면 markers 인잣값에 표현하고자 하는 기호 리스트에 담아 전달
5. lmplot 메서드로 4개의 데이터 그룹에 대한 그래프 한 번에 그리기.
6. 그룹별로 그래프 나눠 그리기 col, col_warp 인자 설정. col_warp: 그래프를 그릴 열의 최댓값 지정 col: 데이터 그룹을 구분할 열 지정

#### FaceGrid 클래스로 그룹별 그래프 그리기
데이터프레임과 그룹을 구분할 열 이름 전달해 객체 생성. 객체의 map 메서드에 그래프 종류, x축 지정할 열 이름 전달해 히스토그램, 밀집도, 양탄자 그래프 그리기

```python
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```python
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel("Frequency")
axes1.set_ylabel("Total Bill")
fig
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title("Scatterplot of Total Bill vs Tip")
axes1.set_xlabel('Total Bill')
axes1.set_ylabel("Tip")
```

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
              labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel("Tip")
axes1.set_title('Boxplot of Tips by Sex')
```

```python
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10,
    c = tips['sex_color'],
    alpha = 0.5
)
axes1.set_title("Total Bill vs Tip Colored by Sex and Sized by Size")
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
import seaborn as sns

tips = sns.load_dataset('tips')
```

```python
import matplotlib.pyplot as plt
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title("Total Bill Histogram with Density Plot")
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel("Frequency")
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
ax = plt.subplots()
ax = sns.countplot(x='day', data=tips)
sns.countplot()
ax.set_title('Count of Days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title("Scatterplot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title("Scatterplot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
ax = plt.subplots()
ax = sns.kdeplot(x=tips['total_bill'],
                 y=tips['tip'],
                 shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total Bill')
```

```python
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title("Violin plot of total bill by time of day")
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
fig = plt.figure()
seaborn_styles=['darkgrid','whitegrid','dark','white','ticks']
for idx, style in enumerate(seaborn_styles):
  plot_position = idx + 1
  with sns.axes_style(style):
    ax = fig.add_subplot(2,3, plot_position)
    violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
    violin.set_title(style)
fig.tight_layout()
```
